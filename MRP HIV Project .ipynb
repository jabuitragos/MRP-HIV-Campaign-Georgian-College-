```python
import pandas as pd

# Here we have our dataset
file_path = "HIV_Prevention_Adolescents_2024.xlsx"
xls = pd.ExcelFile(file_path)

# In the Notes Sheet we have our new names for our key performance indicators
df_notes = pd.read_excel(xls, sheet_name="Notes", header=None)

# Extract old and new indicator names
indicator_mapping = df_notes.iloc[4:, [1, 4]].dropna()
indicator_mapping.columns = ["Old Indicator", "New Indicator"]

# Convert to dictionary to formalize the new names
indicator_rename_dict = dict(zip(indicator_mapping["Old Indicator"], indicator_mapping["New Indicator"]))

# Now loading Notes sheet
df_data = pd.read_excel(xls, sheet_name="Data", header=1)

# Rename the indicators in the Data sheet
df_data["Indicator"] = df_data["Indicator"].replace(indicator_rename_dict)

#Filtering the data to show only data after 2015
filtered_data = df_data[(df_data['Year'] >= 2015) & (df_data['Year'] <= 2023)]
print(filtered_data)

# Save our new dataset
df_data.to_excel("Cleaned_HIV_Prevention_Data.xlsx", index=False)

# Display  first columns
df_data.head()
```
